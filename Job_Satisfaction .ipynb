```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
df = pd.read_csv('Course 1.csv')

# Question 1: Average job satisfaction among respondents with different professional statuses
professional_job_satisfaction = df.groupby('Professional')['JobSatisfaction'].mean().reset_index()
print("Average Job Satisfaction by Professional:")
print(professional_job_satisfaction)

# Plot the average job satisfaction by professional status
plt.figure(figsize=(10, 6))
sns.barplot(x='Professional', y='JobSatisfaction', data=professional_job_satisfaction)
plt.xticks(rotation=90)
plt.title('Average Job Satisfaction by Professional')
plt.xlabel('Professional')
plt.ylabel('Average Job Satisfaction')
plt.show()
```

```python
# Question 2: Job satisfaction among professional developers vs non-developers who sometimes write code
dev_vs_nondev = df[df['Professional'].isin(['Professional developer', 'Professional non-developer who sometimes writes code'])]
dev_nondev_job_satisfaction = dev_vs_nondev.groupby('Professional')['JobSatisfaction'].mean().reset_index()
print("Job Satisfaction among Professional Developers vs Non-Developers:")
print(dev_nondev_job_satisfaction)

# Plot the job satisfaction among professional developers vs non-developers
plt.figure(figsize=(10, 6))
sns.barplot(x='Professional', y='JobSatisfaction', data=dev_nondev_job_satisfaction)
plt.xticks(rotation=90)
plt.title('Job Satisfaction among Professional Developers vs Non-Developers')
plt.xlabel('Professional')
plt.ylabel('Job Satisfaction')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

class JobSatisfactionAnalysis:
    def __init__(self, file_path):
        """
        Initialize the JobSatisfactionAnalysis class with the dataset.

        Parameters:
        file_path (str): The path to the CSV file containing the dataset.
        """
        self.df = pd.read_csv(file_path)
        self.professional_job_satisfaction = None
        self.dev_nondev_job_satisfaction = None
        self.selected_professionals = None

    def calculate_average_job_satisfaction(self):
        """
        Calculate the average job satisfaction for each professional status.
        """
        self.professional_job_satisfaction = self.df.groupby('Professional')['JobSatisfaction'].mean().reset_index()

    def calculate_dev_nondev_job_satisfaction(self):
        """
        Calculate the job satisfaction among professional developers and non-developers who sometimes write code.
        """
        dev_vs_nondev = self.df[self.df['Professional'].isin(['Professional developer', 'Professional non-developer who sometimes writes code'])]
        self.dev_nondev_job_satisfaction = dev_vs_nondev.groupby('Professional')['JobSatisfaction'].mean().reset_index()

    def calculate_selected_professionals(self):
        """
        Select students, professional developers, and non-developers who sometimes write code for analysis.
        """
        self.selected_professionals = self.df[self.df['Professional'].isin(['Student', 'Professional developer', 'Professional non-developer who sometimes writes code'])]

    def plot_average_job_satisfaction(self):
        """
        Plot the average job satisfaction by professional status.
        """
        plt.figure(figsize=(8, 5))
        bar_plot = sns.barplot(x='Professional', y='JobSatisfaction', data=self.professional_job_satisfaction)
        plt.xticks(rotation=90)
        plt.title('Average Job Satisfaction by Professional')
        plt.xlabel('Professional')
        plt.ylabel('Average Job Satisfaction')

        # Add labels inside the bars
        for index, row in self.professional_job_satisfaction.iterrows():
            bar_plot.text(index, row.JobSatisfaction, round(row.JobSatisfaction, 2), color='black', ha="center")

        plt.show()

    def plot_dev_nondev_job_satisfaction(self):
        """
        Plot the job satisfaction among professional developers and non-developers.
        """
        plt.figure(figsize=(8, 5))
        bar_plot = sns.barplot(x='Professional', y='JobSatisfaction', data=self.dev_nondev_job_satisfaction)
        plt.xticks(rotation=90)
        plt.title('Job Satisfaction among Professional Developers vs Non-Developers')
        plt.xlabel('Professional')
        plt.ylabel('Job Satisfaction')

        # Add labels inside the bars
        for index, row in self.dev_nondev_job_satisfaction.iterrows():
            bar_plot.text(index, row.JobSatisfaction, round(row.JobSatisfaction, 2), color='black', ha="center")

        plt.show()

    def plot_selected_professionals_distribution(self):
        """
        Plot the distribution of job satisfaction among students, professional developers, and non-developers.
        """
        plt.figure(figsize=(8, 5))
        sns.boxplot(x='Professional', y='JobSatisfaction', data=self.selected_professionals)
        plt.xticks(rotation=90)
        plt.title('Distribution of Job Satisfaction among Students, Professional Developers, and Non-Developers')
        plt.xlabel('Professional')
        plt.ylabel('Job Satisfaction')
        plt.show()

# Create an instance of the JobSatisfactionAnalysis class
analysis = JobSatisfactionAnalysis('Course 1.csv')

# Calculate the necessary data
analysis.calculate_average_job_satisfaction()
analysis.calculate_dev_nondev_job_satisfaction()
analysis.calculate_selected_professionals()

# Plot the results
analysis.plot_average_job_satisfaction()
analysis.plot_dev_nondev_job_satisfaction()
analysis.plot_selected_professionals_distribution()
    
```
